 # Singly Linked Lists

 ## Michael T. Goodrich, Roberto Tamassia, Michael H. Goldwasser. Data Structures and Algorithms in Java. Wiley 2014. Capítulo 3.2, página 122

```
public class SinglyLinkedList<E> {
    //---------------- nested Node class ----------------
    private static class Node<E> {
        private E element; // head node of the list (or null if empty)
        private Node<E> next; // last node of the list (or null if empty)

        public Node(E e, Node<E> n) {
            element = e;
            next = n;
        }

        public E getElement() { return element; }
        public Node<E> getNext() { return next; }
        public void setNext(Node<E> n) { next = n; }
    }
    //----------- end of nested Node class -----------

   // instance variables of the SinglyLinkedList
    private Node<E> head = null; // head node of the list (or null if empty)
    private Node<E> tail = null; // last node of the list (or null if empty)
    private int size = 0; // number of nodes in the list

    public SinglyLinkedList() { } // constructs an initially empty list

    // access methods
    public int size() { return size; }
    public boolean isEmpty() { return size == 0; }

    public E first() { // returns (but does not remove) the first element
        if (isEmpty()) return null;
        return head.getElement();
    }

    public E last() { // returns (but does not remove) the first element
        if (isEmpty()) return null;
        return tail.getElement();
    }

    // update methods
    public void addFirst(E e) { // adds element e to the front of the list
        head = new Node<>(e, head); // create and link a new node
        if (size == 0)
            tail = head; // special case: new node becomes tail also
        size++;
    }

    public void addLast(E e) { // adds element e to the end of the list
        Node<E> newest = new Node<>(e, null); // node will eventually be the tail
        if (isEmpty())
            head = newest; // special case: previously empty list
        else
            tail.setNext(newest); // new node after existing tail
        tail = newest; //new node becomes the tail
        size++;
    }

    public E removeFirst() { // removes and returns the first element
        if (isEmpty()) return null; // nothing to remove
        E answer = head.getElement();
        head = head.getNext(); // se convertirá en null si la lista tenía solo un nodo
        size--;
        if (size == 0)
            tail = null; // caso especial ya que la lista está ahora vacía
        return answer;
    }

    public E removeLast() { // removes and returns the last element
        if (isEmpty()) return null; // nothing to remove
        E answer = tail.getElement();

        if (size == 1) { // Only one Node
            head = tail = null;
        } else {
            Node<E> current = head;
            while (current.getNext() != tail) {
                current = current.getNext();
            }
            current.setNext(null); // unlinks the tail
            tail = current; // the new tail is the previous node
        }
        size--;
        return answer;
    }

   public void printList() {
        Node<E> current = head;
        while (current != null) {
            System.out.print(current.getElement() + " ");
            current = current.getNext();
        }
        System.out.println();
    }

    public static void main(String[] args) {
        SinglyLinkedList list = new SinglyLinkedList();
        list.addFirst(1);
        list.addLast(2);
        list.addLast(3);
        list.printList(); // Imprime el: 1 2 3
        list.removeFirst();
        list.printList(); // Imprime el: 2 3
        list.removeLast();
        list.printList(); // Imprime el: 2
    }
 }
```

#  Circularly Linked Lists

```
public class CircularlyLinkedList<E> {
    //---------------- nested Node class ----------------
    private static class Node<E> {
        private E element; // referencia al elemento almacenado en este nodo
        private Node<E> next; // referencia al siguiente nodo en la lista

        public Node(E e, Node<E> n) {
            element = e;
            next = n;
        }

        public E getElement() { return element; }
        public Node<E> getNext() { return next; }
        public void setNext(Node<E> n) { next = n; }
    }
    //----------- end of nested Node class -----------

    // variables de instancia de CircularlyLinkedList
    private Node<E> tail = null; // almacenamos el tail (pero no el head)
    private int size = 0; // número de nodos en la lista

    public CircularlyLinkedList() { } // construye una lista inicialmente vacía

    // métodos de acceso
    public int size() { return size; }
    public boolean isEmpty() { return size == 0; }

    public E first() { // devuelve (pero no elimina) el primer elemento
        if (isEmpty()) return null;
        return tail.getNext().getElement(); // el head está *después* del tail
    }

    public E last() { // devuelve (pero no elimina) el último elemento
        if (isEmpty()) return null;
        return tail.getElement();
    }

    // métodos de actualización
    public void rotate() { // rota el primer elemento al final de la lista
        if (tail != null) // si está vacío, no hace nada
            tail = tail.getNext(); // el antiguo head se convierte en el nuevo tail
    }

    public void addFirst(E e) { // añade el elemento e al principio de la lista
        if (size == 0) {
            tail = new Node<>(e, null);
            tail.setNext(tail); // enlaza consigo mismo circularmente
        } else {
            Node<E> newest = new Node<>(e, tail.getNext());
            tail.setNext(newest);
        }
        size++;
    }

    public void addLast(E e) { // añade el elemento e al final de la lista
        addFirst(e); // inserta el nuevo elemento al frente de la lista
        tail = tail.getNext(); // ahora el nuevo elemento se convierte en el tail
    }

    public E removeFirst() { // elimina y devuelve el primer elemento
        if (isEmpty()) return null; // nada que eliminar
        Node<E> head = tail.getNext();
        if (head == tail) {
            tail = null; // debe ser el único nodo restante
        } else {
            tail.setNext(head.getNext()); // elimina “head” de la lista
        }
        size--;
        return head.getElement();
    }

    public void printList() {
        if (isEmpty()) {
            System.out.println("La lista está vacía.");
            return;
        }
        Node<E> current = tail.getNext(); // el head es el siguiente de tail
        do {
            System.out.print(current.getElement() + " ");
            current = current.getNext();
        } while (current != tail.getNext());
        System.out.println();
    }

    public static void main(String[] args) {
        CircularlyLinkedList<Integer> lista = new CircularlyLinkedList<>();
        lista.addFirst(1);
        lista.addLast(2);
        lista.addLast(3);
        lista.printList(); // Imprime: 1 2 3

        lista.removeFirst();
        lista.printList(); // Imprime: 2 3

        lista.addFirst(4);
        lista.printList(); // Imprime: 4 2 3

        lista.rotate();
        lista.printList(); // Imprime: 2 3 4
}
}
```

```
CircularlyLinkedList ListaCircular = new CircularlyLinkedList();
ListaCircular.main(null);
```

# SingleLinked Lists

## [Koffman, Elliot B.;](https://) Wolfgang, Paul A. T. Data structures : abstraction and design using Java. Wiley 2016. Capítulo 2.5, página 77.

```
public class KWSingleLinkedList<E> {
    /** A Node is the building block for a singly-linked list. */
    private static class Node<E> {
        // Data Fields
        /** The reference to the data. */
        private E data;
        /** The reference to the next node. */
        private Node<E> next;

        // Constructors
        /** Creates a new node with a null next field.
         * @param dataItem The data stored
         */
        private Node(E dataItem) {
            data = dataItem;
            next = null;
        }

        /** Creates a new node that references another node.
         * @param dataItem The data stored
         * @param nodeRef The node referenced by the new node
         */
        private Node(E dataItem, Node<E> nodeRef) {
            data = dataItem;
            next = nodeRef;
        }
    }

    /** Reference to list head. */
    private Node<E> head = null;
    /** The number of items in the list */
    private int size = 0;

    /** Add an item to the front of the list.
     * @param item The item to be added
     */
    public void addFirst(E item) {
        head = new Node<>(item, head);
        size++;
    }

    /** Add a node after a given node
     * @param node The node preceding the new item
     * @param item The item to insert
     */
    private void addAfter(Node<E> node, E item) {
        node.next = new Node<>(item, node.next);
        size++;
    }

    /** Remove the node after a given node
     * @param node The node preceding the node to be removed
     * @return The data of the removed node or null if there is no node to remove
     */
    private E removeAfter(Node<E> node) {
        Node<E> temp = node.next;
        if (temp != null) {
            node.next = temp.next;
            size--;
            return temp.data;
        } else {
            return null;
        }
    }

    /** Remove the first node from the list
     * @return The removed node's data or null if the list is empty
     */
    private E removeFirst() {
        Node<E> temp = head;
        if (head != null) {
            head = head.next;
        }
        // Return data at old head or null if list is empty
        if (temp != null) {
            size--;
            return temp.data;
        } else {
            return null;
        }
    }

    /** Find the node at a specified position
     * @param index The position of the node sought
     * @return The node at index or null if it does not exist
     */
    private Node<E> getNode(int index) {
        Node<E> node = head;
        for (int i = 0; i < index && node != null; i++) {
            node = node.next;
        }
        return node;
    }

    /** Get the data at index
     * @param index The position of the data to return
     * @return The data at index
     * @throws IndexOutOfBoundsException if index is out of range
     */
    public E get(int index) {
        if (index < 0 || index >= size) {
            throw new IndexOutOfBoundsException(Integer.toString(index));
        }
        Node<E> node = getNode(index);
        return node.data;
    }

    /** Store a reference to newValue in the element at position index.
     * @param index The position of the item to change
     * @param newValue The new data
     * @return The data previously at index
     * @throws IndexOutOfBoundsException if index is out of range
     */
    public E set(int index, E newValue) {
        if (index < 0 || index >= size) {
            throw new IndexOutOfBoundsException(Integer.toString(index));
        }
        Node<E> node = getNode(index);
        E result = node.data;
        node.data = newValue;
        return result;
    }

    /** Add an item at a specific index
     * @param index The position at which to add the item
     * @param item The item to add
     */
    public void add(int index, E item) {
        if (index < 0 || index > size) {
            throw new IndexOutOfBoundsException(Integer.toString(index));
        }
        if (index == 0) {
            addFirst(item);
        } else {
            Node<E> node = getNode(index - 1);
            addAfter(node, item);
        }
    }

    /** Append item to the end of the list
     * @param item The item to be appended
     * @return true (as specified by the Collection interface)
     */
    public boolean add(E item) {
        add(size, item);
        return true;
    }

    /** Get the size of the list */
    public int size() {
        return size;
    }

    /** Print the contents of the list */
    public void printList() {
        Node<E> current = head;
        while (current != null) {
            System.out.print(current.data + " ");
            current = current.next;
        }
        System.out.println();
    }

    public static void main(String[] args) {
        KWSingleLinkedList<String> names = new KWSingleLinkedList<>();
        names.addFirst("Sam");
        names.addFirst("Harry");
        names.addFirst("Dick");
        names.addFirst("Tom");

        // Print the list
        System.out.println("List after adding names:");
        names.printList();

        // Add "Bob" after "Harry"
        Node<String> harryNode = names.getNode(1); // Get node for "Harry"
        names.addAfter(harryNode, "Bob");

        // Print the list again
        System.out.println("List after adding Bob after Harry:");
        names.printList();

        // Remove the first node
        names.removeFirst();
        System.out.println("List after removing the first node:");
        names.printList();

        // Get the size of the list
        System.out.println("Size of the list: " + names.size());
    }
}
```

```
KWSingleLinkedList lista = new KWSingleLinkedList();
lista.main(null);
```

 # Double‐Linked Lists and Circular Lists

```
public class DoublyLinkedList<E> {
    private static class Node<E> {
        /** The data value. */
        private E data;
        /** The link to the next node. */
        private Node<E> next = null;
        /** The link to the previous node. */
        private Node<E> prev = null;

        /** Construct a node with the given data value.
         * @param dataItem The data value
         */
        private Node(E dataItem) {
            data = dataItem;
        }
    }

    private Node<E> head = null;
    private Node<E> tail = null;

    // Method to add a new node after a specific node
    public void addAfter(Node<E> node, E newData) {
        Node<E> newNode = new Node<>(newData);
        newNode.next = node.next;
        newNode.prev = node;

        if (node.next != null) {
            node.next.prev = newNode;
        } else {
            tail = newNode; // Update tail if we're adding to the end
        }
        node.next = newNode;
    }

    // Method to remove a node
    public void remove(Node<E> node) {
        if (node.prev != null) {
            node.prev.next = node.next;
        } else {
            head = node.next; // Update head if removing the first node
        }

        if (node.next != null) {
            node.next.prev = node.prev;
        } else {
            tail = node.prev; // Update tail if removing the last node
        }
    }

    // Example usage of the linked list
    public static void main(String[] args) {
        DoublyLinkedList<String> list = new DoublyLinkedList<>();

        // Creating nodes for demonstration
        Node<String> sam = new Node<>("Sam");
        Node<String> harry = new Node<>("Harry");

        // Setting up the list initially
        list.head = sam;
        sam.next = harry;
        harry.prev = sam;
        list.tail = harry;

        // Adding a new node "Sharon" after "Sam"
        Node<String> sharon = new Node<>("Sharon");
        list.addAfter(sam, "Sharon");

        // Print the list
        printList(list);

        // Removing "Harry"
        list.remove(harry);
        System.out.println("After removing Harry:");
        printList(list);

        // Adding a new head "Tamika"
        Node<String> tamika = new Node<>("Tamika");
        tamika.next = list.head;
        list.head.prev = tamika;
        list.head = tamika;

        System.out.println("After adding Tamika at the head:");
        printList(list);

        // Adding "Shakira" after Tamika
        Node<String> shakira = new Node<>("Shakira");
        shakira.prev = list.head;
        shakira.next = list.head.next;
        if (list.head.next != null) {
            list.head.next.prev = shakira;
        }
        list.head.next = shakira;

        System.out.println("After adding Shakira after Tamika:");
        printList(list);
    }

    // Utility method to print the list
    private static <E> void printList(DoublyLinkedList<E> list) {
        Node<E> current = list.head;
        while (current != null) {
            System.out.print(current.data + " ");
            current = current.next;
        }
        System.out.println();
    }
}
```

```
DoublyLinkedList lista = new DoublyLinkedList();
lista.main(null);
```

# Comparación entre SinglyLinkedList, CircularlyLinkedList y Double‐Linked Lists and Circular Lists

## **Introducción**
Las listas enlazadas son estructuras de datos fundamentales que permiten almacenar elementos de manera dinámica. Este documento compara tres tipos de listas enlazadas: **SinglyLinkedList**, **CircularlyLinkedList** y **Double‐Linked Lists and Circular Lists**.

## **Implementaciones**

### **SinglyLinkedList**
La `SinglyLinkedList` es una lista enlazada simple donde cada nodo tiene un elemento y un puntero(Nodo) al siguiente nodo. Solo se puede recorrer en una dirección (hacia adelante).

### **CircularlyLinkedList**
La `CircularlyLinkedList` es similar a la lista enlazada simple, pero el último nodo apunta de nuevo al primero, formando un ciclo. Esto permite recorridos continuos.

### **Double‐Linked Lists and Circular Lists**
La `Double‐Linked Lists and Circular Lists` Las listas doblemente enlazadas son estructuras de datos formadas por nodos que contienen un dato y dos punteros(Nodo): uno al siguiente nodo y otro al anterior, lo que permite navegar en ambas direcciones y realizar inserciones y eliminaciones de forma eficiente. Por otro lado, las listas circulares conectan el último nodo de la lista con el primero, formando un ciclo, lo que permite recorridos continuos sin un final natural; pueden ser simples o doblemente enlazadas. Ambas estructuras son flexibles y útiles en diversas aplicaciones, y la combinación de ambas en una lista circular doblemente enlazada ofrece navegación bidireccional y un recorrido interminable, facilitando un acceso dinámico a los elementos.

## **Comparación de Ventajas y Desventajas**
---
### **SinglyLinkedList**

#### **Ventajas**
- **Simplicidad**: Implementación sencilla y fácil de entender.
- **Menor uso de memoria**: Cada nodo requiere solo un puntero(Nodo) adicional.
- La ventaja principal de una Singly Linked List es que permite una inserción y eliminación eficiente de elementos en cualquier posición de la lista. (por índice), ya que para acceder a un elemento específico, debemos recorrer la lista desde el inicio hasta el nodo deseado.

#### **Desventajas**
- **Acceso unidireccional**: Solo se puede recorrer hacia adelante.
- **Eliminación del último nodo**: Requiere recorrer la lista, lo que implica un tiempo O(null).

### **CircularlyLinkedList**

#### **Ventajas**
- **Recorrido continuo**: Permite que los elementos se recorran en un ciclo, útil para ciertas aplicaciones.
- **Eficiencia en rotación**: La operación de rotar el primer elemento es directa y eficiente.

#### **Desventajas**
- **Complejidad de implementación**: Manejar el ciclo puede complicar el código.
- **Posibles errores en la lógica**: La naturaleza circular puede dar lugar a errores difíciles de depurar.
- **Rendimiento en ciertas operaciones**: Algunas operaciones, como la búsqueda de un elemento específico, pueden requerir más tiempo si no hay un punto claro de finalización (como el final de una lista), lo que podría llevar a un recorrido completo en el peor de los casos.

### **DoublyLinkedList**

#### **Ventajas**
- **Recorridos bidireccionales**: Permite recorrer la lista tanto hacia adelante como hacia atrás.
- **Eliminación eficiente**: La eliminación de nodos es más eficiente porque no se necesita recorrer toda la lista para encontrar el nodo anterior.
- **Navegación bidireccional**: Se puede recorrer hacia adelante y hacia atrás.

#### **Desventajas**
- **Mayor uso de memoria**: Cada nodo requiere dos punteros(Nodo), aumentando el uso de memoria.
- **Mayor complejidad**: La implementación es más compleja en comparación con las listas enlazadas simples.
- **Problemas de coherencia**: En un entorno multihilo, mantener la coherencia de las referencias puede ser más desafiante, lo que podría requerir mecanismos de sincronización más complejos.

# **Diferencias**
## **Estructura**:

- **Lista Doblemente Enlazada**: Cada nodo tiene punteros al siguiente y al anterior.
- **Lista Circular**: El último nodo apunta de nuevo al primero; puede ser simple o doble.
- **Lista Enlazada Simple**: Cada nodo tiene solo un puntero al siguiente nodo.
## **Dirección**:

- **Lista Doblemente Enlazada**: Bidireccional (puede navegar hacia adelante y hacia atrás).
- **Lista Circular**: Puede ser bidireccional o unidireccional, dependiendo de su implementación.
- **Lista Enlazada Simple**: Unidireccional (solo hacia adelante).
## **Uso de Memoria**:

- **Lista Doblemente Enlazada**: Mayor uso de memoria por los dos punteros por nodo.
- **Lista Circular**: Uso de memoria variable; depende de si es simple o doble.
- **Lista Enlazada Simple**: Menor uso de memoria con un solo puntero por nodo.
## **Complejidad de Operaciones**:

- **Lista Doblemente Enlazada**: Más compleja debido a la gestión de dos punteros.
- **Lista Circular**: Mayor complejidad al mantener la estructura circular.
- **Lista Enlazada Simple**: Más sencilla de implementar y usar, pero menos flexible.

##  **Conclusión**
Cada tipo de lista enlazada tiene sus propias ventajas y desventajas. La elección entre `SinglyLinkedList`, `CircularlyLinkedList` y `DoublyLinkedList` dependerá de los requisitos específicos de la aplicación, como la necesidad de recorridos unidireccionales o bidireccionales, la simplicidad en la implementación, o la eficiencia en operaciones específicas.
